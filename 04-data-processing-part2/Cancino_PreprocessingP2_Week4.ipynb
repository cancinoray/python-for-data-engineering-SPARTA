### Removing the Outlier

```python
newdata.describe(include = 'all' )
```

```python
m = np. mean(newdata.Price)
s = np.std(newdata.Price)

print (m)
print(s)

ll = max(0 , m -3*s)
print (ll)

ul = m + 3*s
print (ul)
```

```python
newdata_cleaned1 = newdata[newdata.Price >= ll]
newdata_cleaned2 = newdata_cleaned1[newdata_cleaned1.Price <= ul]
```

```python
newdata_cleaned2.describe(include = 'all' )
```

### Standardizing the Price column

```python
newdata_cleaned2.insert(4 , 'Standardized' , 0)
```

```python
###Standardization Formula: Xnew = (Xold - mean)/ standard deviation

newdata_cleaned2.Standardized = (newdata_cleaned2.Price - np.mean(newdata_cleaned2.Price))/(np.std(newdata_cleaned2.Price))
```

### Part 2

```python
data = pd.read_csv("Cancino_Handling_Missing_Values.csv")
```

```python
data
```

```python
data.isnull().sum()
```

```python
## Columns with missing data
## Define null columns
null_columns = data.columns[data.isnull().any()]

## Print columns containing null values
data[data.isnull().any(axis = 1)][null_columns]
```

```python
## I choose to delete the rows that containing NaN values more than 2
data = data.dropna(thresh = 6,  axis=0)
data
```

```python
## Columns with missing data
## Define null columns
null_columns = data.columns[data.isnull().any()]

## Print columns containing null values
data[data.isnull().any(axis = 1)][null_columns]
```

```python
#Data Imputation
#Data Imputation: Mean
#Replacing missing values using mean

mean_value = data['Income'].mean() #getting the mean

print(mean_value)

data['Income'] = data['Income'].fillna(mean_value) #fill mean values

# For me, it's logical to use mean to fill out the NaN values in Income column,as comparing the relationship between the income and the age

data
```

```python
#Data Imputation
#Data Imputation: Mode
#Replacing missing values using mode

mode_value = data['Sex'].mode() #getting the mode

print(mode_value)

data['Sex'] = data['Sex'].fillna(mode_value[0]) #fill mode values
data

#I use the mode because we are dealing with categorical values.
```

```python
#Data Imputation
#Data Imputation: Median
#Replacing missing values using mode

median_value = data['Children'].median() #getting the median

print(median_value)

data['Children'] = data['Children'].fillna(median_value) #fill median values
data

#in the data it is logical to use the median to fill out the NaN values
```
